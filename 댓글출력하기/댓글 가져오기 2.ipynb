```python
# selenium.webdriver의 chrome을 import해 url를 driver.get.
import selenium
from selenium.webdriver import chrome
driver.get('https://n.news.naver.com/mnews/ranking/article/comment/092/0002279569?sid=001')
```

```python
# common.by에서 by를 import
import time
from selenium.webdriver.common.by import By
for i in range(3):
    time.sleep(1);
    driver.find_element(By.CLASS_NAME, 'u_cbox_btn_more').click()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'lxml').find_all('span', {'class':'u_cbox_contents'})
for i in soup:
    print(i.text+"\n")
```
