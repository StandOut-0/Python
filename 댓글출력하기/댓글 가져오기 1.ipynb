```python
from selenium.webdriver.common.by import By
```

```python
driver.find_elements(By.CLASS_NAME, 'u_cbox_btn_more')
```

```python
for i in range(10):
    li = driver.find_elements(By.CLASS_NAME, 'u_cbox_btn_more')
    for i in li:
        print(i.text)
```
