```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = 'https://www.dailian.co.kr/news/view/1190364/?sc=Naver'
soup = BeautifulSoup(urlopen(url).read(), "lxml")
print(soup)
```

```python
print(soup.title.get_text())
print(soup.find("h2","title").text+"\n\n")
print(soup.find("div","inner-subtitle").text+"\n\n")
soup.find("div","flexwrap").decompose() # 쓸데없는 #하이퍼링크를 삭제했다.
print(soup.find("div","article").text+"\n\n")
```
